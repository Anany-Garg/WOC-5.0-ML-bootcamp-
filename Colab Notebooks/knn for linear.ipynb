```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dfp=pd.read_csv('/content/drive/MyDrive/woc files/linear_train.csv')
dft=pd.read_csv('/content/drive/MyDrive/woc files/linear_test_data.csv')
```

```python
x_test1=np.array(dft.iloc[:,1:])
print(dfp)
x_train=np.array(dfp.iloc[:,1:-1])
y_train=np.array(dfp.iloc[:,-1])
print(x_train.shape)
y_train.shape
```

```python
mu=np.mean(x_train,axis=0)
sigma=np.std(x_train,axis=0)
x_train_norm=((x_train-mu)/sigma)
```

```python
fu=np.random.rand(x_train_norm.shape[0])
fuo=fu<0.8
fua=fu>0.8
x_train_train=x_train_norm[fuo,:]
x_test=x_train_norm[fua,:]
y_train_train=y_train[fuo]
y_test=y_train[fua]

```

```python
x_test_norm=((x_test1-mu)/sigma)

Xtest=x_test_norm

Xt = x_train_train
Xte=x_test
yte=y_test
yt = y_train_train
mt,n=Xt.shape
print(Xte.shape)
mte=Xte.shape[0]
```

```python
def p(X,Xeu,y,k):
  mt,n=Xt.shape
  Xe=np.reshape(Xeu,(-1,n))
  d1=(np.matmul(X,Xe.T)*(-2))
  d2=np.sum((X**2),axis=1).reshape(-1,1)
  d3=np.sum((Xe**2),axis=1).reshape(1,-1)
  d4=np.sqrt((d1+d2+d3))
  d5=np.argsort(d4,axis=0)
  d6=y[d5[:k,:]] 
  d7=np.mean(d6,axis=0)
  return(d7)
```

```python
for i in range (15,25):
  pre=p(Xt,Xte,yt,i)
  TSS=np.sum((yt-np.mean(yt))**2)
  RSS=np.sum((pre-yte)**2)
  R_2=(1-(RSS/TSS))
  print(R_2)
```
